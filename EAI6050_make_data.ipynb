```python
data_dict = {}
for i in range(len(df_1)):
  rec = df_1.iloc[i]
  this_day = rec.date
  prev_date = pd.Timestamp(this_day.date() - timedelta(days = 1))

  price_today = float(df_2.loc[df_2.date == this_day].high)
  price_yesterday = float(df_2.loc[df_2.date == prev_date].high)

  news_articles = rec.news

  data_dict[str(this_day.date())] = {'date': str(this_day.date()), 'high': price_today, 'high_yesterday': price_yesterday, 'news': news_articles}
```
